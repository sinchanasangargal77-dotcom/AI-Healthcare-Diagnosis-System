```python
specialist_dict = {

    "Fungal infection": "Dermatologist",

    "Allergy": "Allergist",

    "GERD": "Gastroenterologist",

    "Chronic cholestasis": "Hepatologist",

    "Drug Reaction": "General Physician",

    "Peptic ulcer diseae": "Gastroenterologist",

    "AIDS": "Infectious Disease Specialist",

    "Diabetes ": "Endocrinologist",

    "Bronchial Asthma": "Pulmonologist"
}
```

```python
def recommend_specialist(disease):

    return specialist_dict.get(disease, "General Physician")
```

```python
disease = predict_disease(sample)

doctor = recommend_specialist(disease)

print("Predicted Disease:", disease)

print("Recommended Specialist:", doctor)
```

## Risk Prediction Function

```python
def predict_risk(symptoms):

    prediction = model.predict(symptoms)

    confidence = prediction.max()

    if confidence > 0.8:
        return "High Risk"

    elif confidence > 0.5:
        return "Moderate Risk"

    else:
        return "Low Risk"
```

```python
risk = predict_risk(sample)

print("Risk Level:", risk)
```

```python
disease = predict_disease(sample)

doctor = recommend_specialist(disease)

risk = predict_risk(sample)

print("Predicted Disease:", disease)

print("Recommended Specialist:", doctor)

print("Risk Level:", risk)
```
